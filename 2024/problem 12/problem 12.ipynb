```python
# Open the file
from collections import defaultdict, deque, Counter, namedtuple
from functools import reduce, cache, cmp_to_key
from heapq import heappush, heappop, heapify
from itertools import permutations, combinations, product
from math import gcd, sqrt, factorial, ceil, floor
from sortedcontainers import SortedDict, SortedList, SortedSet
from pyrival import * # python competitive programming library
from typing import TypeAlias
import re
import sys
sys.setrecursionlimit(10000)
def scan_nums(line):
    # Use a regular expression to find numbers
    numbers = re.findall(r'-?\d+\.?\d*', line)
    # Convert to appropriate numerical types
    numbers = [float(num) if '.' in num else int(num) for num in numbers]
    return numbers

input_file = 'input.txt'
test_file = 'test.txt'

with open(input_file, 'r') as file:
    lines  = file.readlines()
    grid = [[x for x in line.strip()] for line in lines]


directions = [(-1, 0), (1, 0), (0, -1), (0, 1)]

def compute_perimeter(cur_set):
    cur_perimeter = 0
    for r, c in cur_set:
        for dr, dc in directions:
            nr, nc = r + dr, c + dc
            if (nr, nc) not in cur_set:
                cur_perimeter += 1
    return cur_perimeter

def compute_price(grid):

    R, C = len(grid), len(grid[0])
    
    price = 0
    visited = set()
    for cur_r in range(R):
        for cur_c in range(C):
            if (cur_r, cur_c) not in visited:
                cur_type = grid[cur_r][cur_c]
                cur_set = set()
                q = deque([(cur_r, cur_c)])
                #print(f'\n starting region with type {cur_type} ')
                while len(q) > 0:
                    r, c = q.popleft()
                    if (r, c) in visited:
                        continue
                    visited.add((r, c))
                    cur_set.add((r, c))
                    for dr, dc in directions:
                        nr, nc = r + dr, c + dc
                        if 0 <= nr < R and 0 <= nc < C and grid[nr][nc] == cur_type and (nr, nc) not in visited:
                            q.append((nr, nc))
                #print(f'the region with type {cur_type} is has {compute_num_sides(cur_set)} sides')
                cur_area = len(cur_set)
                cur_perimeter = compute_perimeter(cur_set)
                price += cur_area * cur_perimeter
  
    return price

print(compute_price(grid))
            
            
            
    
```

```python

def get_regions(grid):
    R, C = len(grid), len(grid[0])
    visited = set()
    regions = []
    region_areas = []
    for cur_r in range(R):
        for cur_c in range(C):
            if (cur_r, cur_c) not in visited:
                cur_type = grid[cur_r][cur_c]
                cur_set = set()
                q = deque([(cur_r, cur_c)])
                #print(f'\n starting region with type {cur_type} ')
                while len(q) > 0:
                    
                    r, c = q.popleft()
                    if (r, c) in visited:
                        continue
                    visited.add((r, c))
                    cur_set.add((r, c))
                    for dr, dc in directions:
                        nr, nc = r + dr, c + dc
                        if 0 <= nr < R and 0 <= nc < C and grid[nr][nc] == cur_type and (nr, nc) not in visited:
                            q.append((nr, nc))
                #print(f'the region with type {cur_type} is has {compute_num_sides(cur_set)} sides')
                regions.append(cur_set)
                region_areas.append(len(cur_set))
    return regions, region_areas
    

def group(nums):
    if not nums:
        return 0
    n = len(nums)
    i = 0
    count = 0
    while i < n - 1:
        cur = nums[i]
        upcoming = nums[i + 1]
        if cur == upcoming - 1:
            i += 1
        else:
            count += 1
            i += 1
            
    count += 1
    
    return count
                
            
        
        
def compute_num_sides(region):
    # columns dict 
    # rows dict
    column_dict = defaultdict(list)
    row_dict = defaultdict(list) 
    for r, c in region:
        column_dict[c].append(r)
        row_dict[r].append(c)
    for c in column_dict:
        column_dict[c].sort()
    for r in row_dict:
        row_dict[r].sort()
    sides = 0
    # up and down
    for r in row_dict:
        cur_row = row_dict[r]
        up_edges = [c for i, c in enumerate(cur_row) if (r - 1, c) not in region]
        down_edges = [c for i, c in enumerate(cur_row) if (r + 1, c) not in region]
        sides += group(up_edges) + group(down_edges)

    for c in column_dict:
        cur_column = column_dict[c]
        left_edges = [r for i, r in enumerate(cur_column) if (r, c - 1) not in region]
        right_edges = [r for i, r in enumerate(cur_column)  if (r, c + 1) not in region]
        sides += group(left_edges) + group(right_edges)
    return sides

      
with open(input_file, 'r') as file:
    lines  = file.readlines()
    grid = [[x for x in line.strip()] for line in lines]        
        
def compute_discount_price(grid):
    regions, region_areas = get_regions(grid)
    
    price = 0
    for i, region in enumerate(regions):
        num_sides = compute_num_sides(region)
        region_area = region_areas[i]
        price += region_area * num_sides
    return price

print(compute_discount_price(grid))
```
