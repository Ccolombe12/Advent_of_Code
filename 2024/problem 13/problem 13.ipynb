```python
from pyomo.environ import ConcreteModel, Var, Objective, Constraint, Integers, NonNegativeIntegers, SolverFactory

# Define given constants
x1 = 2
x2 = 3
y1 = 1
y2 = 4
X = 8
Y = 10

# Create a concrete model
model = ConcreteModel()

# Define decision variables
# Both a and b are non-negative integers
model.a = Var(domain=NonNegativeIntegers)
model.b = Var(domain=NonNegativeIntegers)

# Objective function: Minimize 3a + b
model.obj = Objective(expr=3*model.a + model.b, sense=1)  # 1 means minimize

# Constraints
model.const1 = Constraint(expr=model.a*x1 + model.b*x2 == X)
model.const2 = Constraint(expr=model.a*y1 + model.b*y2 == Y)

# Choose a solver
# Note: You must have a suitable MILP solver installed. 
# For example, if using CBC:
solver = SolverFactory('cbc')

# Solve the model
results = solver.solve(model, tee=True)

# Print the results
if (results.solver.status == 'ok') and (results.solver.termination_condition == 'optimal'):
    print("Optimal solution found!")
    print("a =", model.a())
    print("b =", model.b())
    print("Objective value =", model.obj())
else:
    print("No optimal solution found. Solver Status:", results.solver.status)
    print("Termination Condition:", results.solver.termination_condition)
```
