```python
# Open the file
from collections import defaultdict, deque, Counter, namedtuple
from functools import reduce, lru_cache, cmp_to_key
from heapq import heappush, heappop, heapify
from itertools import permutations, combinations, product
from math import gcd, sqrt, factorial, ceil, floor
from sortedcontainers import SortedDict, SortedList, SortedSet
import re
import sys
sys.setrecursionlimit(10000)

def scan_nums(line):
    # Use a regular expression to find numbers
    numbers = re.findall(r'-?\d+\.?\d*', line)
    # Convert to appropriate numerical types
    numbers = [float(num) if '.' in num else int(num) for num in numbers]
    return numbers

input_file = 'input.txt'
test_file = 'test.txt'
# ----------------------------------------------------------------------------
with open(input_file, 'r') as file:
    disk = file.readline().strip()
    
# construct the file
def get_disk_rep(disk):
    """Gives the uncompressed representation of the disk"""
    disk_rep = []
    id = 0
    for i, c in enumerate(disk):
        c = int(c)
        if i % 2 == 0:
            disk_rep.extend([id for _ in range(c)])
            id += 1
        else:
            disk_rep.extend(['.' for _ in range(c)])
    return disk_rep



def empty_pos(arr):
    """ Gives a deque of the positions of the empty positions"""
    empty_idx = deque([i for i, c in enumerate(arr) if c == '.'])
    return empty_idx
    
def shifted_disk_rep(disk_rep):
    """Gives the resulting disk rep after removing left empty cells"""
    n  = len(disk_rep)
    empty_positions = empty_pos(disk_rep)
    for r in range(n - 1, -1, -1):
        if disk_rep[r] == '.':
            continue
        if not empty_positions:
            break
        l = empty_positions.popleft()
        if l >= r:
            break
        disk_rep[r], disk_rep[l] = disk_rep[l], disk_rep[r]
        
    return disk_rep
    

def compute_ans(disk_rep):
    """Computes the checksum"""
    S = 0
    for i, c in enumerate(disk_rep):
        if c == '.':
            break
        S += i * int(c)
    return S

final_disk_rep = shifted_disk_rep(get_disk_rep(disk))
print(compute_ans(final_disk_rep))
        
        

    
```

```python
with open(input_file, 'r') as file:
    
    disk = file.readline().strip()
# create the file we are going to use


# can we keep track of the size of the  empty spaces?

# store, (size, start, end)
# also keep track of where the blocks are

# I think we can just scan l -> r when looking for a place to put a block

def get_block_left(i, arr, t):
    s = i
    while i >= 0 and arr[i] == t:
        i -= 1
    e = i
    return e + 1, s  

def get_block_right(i, arr, t):
    n = len(arr)
    s = i
    while i < n and arr[i] == t:
        i += 1
    e = i 
    return s, e - 1

disk_rep = get_disk_rep(disk)
# find the empty blocks
empty_blocks = SortedList()
i = 0
while i < n:
    c = disk_rep[i]
    if c == '.':
        s, e = get_block_right(i, disk_rep, '.')
        empty_blocks.add((s, e))
        i = e + 1
       
    else:
        i += 1

# traverse the file trying to move rightmost block.
r = n - 1
while r > 0:
    
    if disk_rep[r] == '.':
        r -= 1
    else:
        #print(f' at {r = } and we see {file[r] = }')
        id = disk_rep[r]
        s_b, e_b = get_block_left(r, disk_rep, id)
        
        len_id = e_b - s_b + 1
        #print(f'looking at the block of {id}"s at  from {s_b} to {e_b},there are {len_id} of them ')
        
        for i, (s, e) in enumerate(empty_blocks):
            len_block = e - s + 1
            #print(f'the empty block from {s} to {e} is {len_block} long')
            if len_block >= len_id and e < s_b:
               
                # update the file: move block into gap
                for i in range(s, s + len_id):
                    disk_rep[i] = id
                for j in range(s_b, e_b + 1):
                    disk_rep[j] = '.'
                    
                # update the empty blocks
                empty_blocks.remove((s, e))
                empty_blocks.add((s + len_id, e))
                
                # update pointer 
                r = s_b - 1   
                break
        # this block can't fit in any of the empty blocks. It must stay   
        r = s_b - 1
            

S = 0
for i, c in enumerate(disk_rep):
    if c != '.':
        S += i * int(c)
print(S)

        
        




    
    
```
