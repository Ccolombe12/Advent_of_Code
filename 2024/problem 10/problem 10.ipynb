```python
# Open the file
from collections import defaultdict, deque, Counter, namedtuple
from functools import reduce, lru_cache, cmp_to_key
from heapq import heappush, heappop, heapify
from itertools import permutations, combinations, product
from math import gcd, sqrt, factorial, ceil, floor
from sortedcontainers import SortedDict, SortedList, SortedSet
import re
import sys
sys.setrecursionlimit(10000)

def scan_nums(line):
    # Use a regular expression to find numbers
    numbers = re.findall(r'-?\d+\.?\d*', line)
    # Convert to appropriate numerical types
    numbers = [float(num) if '.' in num else int(num) for num in numbers]
    return numbers

input_file = 'input.txt'
test_file = 'test.txt'
# ----------------------------------------------------------------------------
with open(input_file, 'r') as file:
    grid = [[int(x) for x in line.strip()] for line in file]

def get_trailhead_loc(grid):
    R, C = len(grid), len(grid[0])
    trailhead_loc = [(r, c) for r in range(R) for c in range(C) if grid[r][c] == 0]
    return trailhead_loc

directions = [(0,1), (1,0), (-1,0), (0,-1)]
# for each trailhead, count number of "paths" to a peak (note path cant be longer than 9)
# now traverse paths to look for peaks. Just do a bfs

def score_trailhead_1(trailhead_loc, grid):
    R, C = len(grid), len(grid[0])
    score = 0
    r, c = trailhead_loc
    visited = set()
    # no need for previous since path is directed by inc. of height
    q = deque([(r, c)])
    while q:
        r, c = q.popleft()
        if (r, c) in visited:
            continue
        visited.add((r, c))
        if grid[r][c] == 9:
            score += 1
            continue
        
        for dr, dc in directions:
            nr, nc = r + dr, c + dc
            if 0 <= nr < R and 0 <= nc < C and grid[nr][nc] == grid[r][c] + 1:
                q.append((nr, nc))
    return score
def compute_ans_1(trailhead_loc, grid):
    total = 0
    for trailhead in trailhead_loc:
        score = score_trailhead_1(trailhead, grid)
        #print(f' {trailhead}: {score}')
        total += score
    return total

trailhead_loc = get_trailhead_loc(grid)
print(compute_ans_1(trailhead_loc, grid))

                
    
    




```

```python
# now we need to count the distinct paths from each trailhead to a peak

# now we need to be able to peel back the path?
# just save the path we use and check if we have already seen it when we hit a peak
# if we have, skip it
with open(input_file, 'r') as file:
    grid = [[int(x) for x in line.strip()] for line in file] 


def score_trailhead_2(trailhead_loc, grid):
    R, C = len(grid), len(grid[0])
    """ Now now just keep track of when we hit a peak and count the number of 
    times we hit one. Only works bc cycles are not allowed. """
    
    r, c = trailhead_loc
    num_paths = [0]
    def explore(r, c):
        if grid[r][c] == 9:
            num_paths[0] += 1
            return 
        else:
            for dr, dc in directions:
                nr, nc = r + dr, c + dc
                if 0 <= nr < R and 0 <= nc < C and grid[nr][nc] == grid[r][c] + 1:
                    explore(nr, nc)
            return 
    explore(r, c)
    return num_paths[0]

def compute_solution_2(trailhead_loc, grid):
    total = 0
    for trailhead in trailhead_loc:
        score = score_trailhead_2(trailhead, grid)
        total += score
        #print(f' {trailhead}: {score}')
    return total

trailhead_loc = get_trailhead_loc(grid)
print(compute_solution_2(trailhead_loc, grid))
    




```
